```python
final=[]
for i in df["Country Name"].unique():
    df1=df[df["Country Name"]==i]
    data=df1.values
    gdp1=[0]
    for i in range(1,len(data)):
        prev=data[i-1][3]
        cur=data[i][3]
        gdp1.append(round((cur-prev)/prev*100,2))
    df1=df1.assign(GDP=gdp1)
    final.append(df1)
    


    
```

```python
df=pd.concat(final,axis=0)
```
